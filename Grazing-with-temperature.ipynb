```python
import xarray as xr
import numpy as np
from matplotlib import pyplot as plt
```

```python
rootdir = '/archive/oar.gfdl.cmip6/ESM4/DECK/ESM4_piControl_D/gfdl.ncrc4-intel16-prod-openmp/pp/'
datadir_temperature = 'ocean_monthly_z/ts/monthly/5yr/'
filename_temperature = 'ocean_monthly_z.071101-071512.thetao.nc'

datadir_grazing = 'ocean_cobalt_omip_tracers_month_z/ts/monthly/5yr/'
filename_grazing = 'ocean_cobalt_omip_tracers_month_z.071101-071512.graz.nc'


temperature = xr.open_dataset(rootdir+datadir_temperature+filename_temperature)
grazing = xr.open_dataset(rootdir+datadir_grazing+filename_grazing)
```

```python
print(temperature.thetao)
```

```python
print(grazing.graz)
```

```python
ds = xr.merge([temperature, grazing])


ds = ds.set_coords('thetao')

 
theta_bins = np.arange(-2,30,1) 
grazing_in_theta = ds.graz.isel(time=5).groupby_bins('thetao',theta_bins)
```

```python
grazing_in_theta.count(xr.ALL_DIMS).plot()
```

```python
grazing_in_theta.mean(xr.ALL_DIMS).plot()
```
