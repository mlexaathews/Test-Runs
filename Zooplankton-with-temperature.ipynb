```python
## Let's look at how zooplankton are distributed throughout
## the ocean based on temperature!
```

```python
import xarray as xr
import numpy as np
from matplotlib import pyplot as plt
```

```python
rootdir = '/archive/oar.gfdl.cmip6/ESM4/DECK/ESM4_piControl_D/gfdl.ncrc4-intel16-prod-openmp/pp/'
datadir_temperature = 'ocean_monthly_z/ts/monthly/5yr/'
filename_temperature = 'ocean_monthly_z.071101-071512.thetao.nc'

datadir_zooplankton = 'ocean_cobalt_omip_tracers_month_z/ts/monthly/5yr/'
filename_zooplankton = 'ocean_cobalt_omip_tracers_month_z.071101-071512.zooc.nc'


temperature = xr.open_dataset(rootdir+datadir_temperature+filename_temperature)
zooplankton = xr.open_dataset(rootdir+datadir_zooplankton+filename_zooplankton)
```

```python
print(temperature.thetao)
```

```python
print(zooplankton.zooc)
```

```python
ds = xr.merge([temperature, zooplankton])


ds = ds.set_coords('thetao')

 
theta_bins = np.arange(-2,30,1) 
zooplankton_in_theta = ds.zooc.isel(time=5).groupby_bins('thetao',theta_bins)
```

```python
zooplankton_in_theta.count(xr.ALL_DIMS).plot()
```

```python
zooplankton_in_theta.mean(xr.ALL_DIMS).plot()
```
